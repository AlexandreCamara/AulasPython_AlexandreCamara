```python
###############################################################################################################################################################################################################################
#####################################################################################################Treinamento do Método Monte Carlo ########################################################################################
##################################### Autor: Alexandre de Resende Camara ######################################################################################################################################################
##################################### Data: 27/02/2021 ########################################################################################################################################################################
#######################################################################################################################################################################################################################
###############################################################################################################################################################################################################################
# Calcule a velocidade final de um objeto em queda livre a partir de 3 metros de altura (sem resistencia do ar). Calcule o tempo que esse objeto demora para cair.
#
###############################################################################################################################################################################################################################
import math
###############################################################################################################################################################################################################################
#
####################################################################################################################################################################
h_0 = 3.0 #posição inicial, em m
h = 0 #posição final, em m
Delta_h = h - h_0
g = 9.8 # ac grav m/s^2
#
t = (-2*Delta_h/g)**(1/2)
v = g*t
##########################################################################################################################################################################
print("t = {:2.2f}".format(t), "s")
print("v = {:2.2f}".format(v), "m/s")
#######################################################################################################################################################################
```
