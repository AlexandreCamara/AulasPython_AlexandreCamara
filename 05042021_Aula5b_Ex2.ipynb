```python
#############################################################################################################################################
#############################################################################################################################################
##################################### Autor: Alexandre de Resende Camara ####################################################################
##################################### Data: 05/04/2021 ######################################################################################
##################################### Nome: Aula 5b, Exercício 2 ###############################################################
#############################################################################################################################################
# Desenhar polígonos de n lados com triângulos inscritos.
#
#############################################################################################################################################
# INÍCIO DO CÓDIGO
#############################################################################################################################################
# 
import math
import turtle 
#############################################################################################################################################
def triangulo(t,length, angle):  #função que escreve um triângulo dependente do tamanho do lado e do ângulo externo
    y=length * math.sin((angle) * math.pi / 180) # lados dos triangulos inscritos no polígono de n lados 
    t.speed(1)
    t.right(angle)
    t.forward(length)
    t.left(90+angle)
    t.forward(2*y)             #leva a tartaruga até o centro geométrico da figura
    t.left(90+angle)
    t.forward(length)
    t.left(180-angle)
    
#############################################################################################################################################
def polygon(t, length, n): #função que escreverá um polígono de n lados com triângulos inscritos
    angle = 360/n
    for i in range (n):
        triangulo(t, length, angle/2)
        t.left(angle)
#############################################################################################################################################
def multipoly (t, length, n):  # função que deslocará a posição da tartaruga para que não haja superposição dos polígonos desenhados
    polygon(t, length, n)
    t.penup()
    t.forward(2*length)
    t.pendown()
#############################################################################################################################################
joana = turtle.Turtle() # Cria uma tartaruga, atribui a t
joana.shape("turtle")   # formato de tartaruga.
joana.color("black") 
joana.width(3)
#############################################################################################################################################
multipoly(joana,100,5)
multipoly(joana,100,6)
multipoly(joana,100,10)
#############################################################################################################################################
# FIM DO CÓDIGO














```
