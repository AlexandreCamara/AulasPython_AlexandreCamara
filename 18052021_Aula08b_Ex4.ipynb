```python
""" PARTE GRÁFICA"""
plt.plot(tempo,posição)
plt.xlabel("tempo (s)",fontsize=15)
plt.ylabel("posição (m)",fontsize=15)

plt.grid(False)
plt.xlim(0, 10)
plt.ylim()
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.rcParams['figure.figsize'] = (15,10)
plt.show()   
```
